## Sumatoria mediante ciclo
Realice la sumatoria de los números desde 1 hasta el 100 usando el ciclo for

```python
suma = 0

for n in range(1, 101, 1):
    suma += n
    
print ("La suma es: ", suma)
```

## Multiplicación mediante sumatoria

Calcule la multiplicación de dos números usando sumas

```python
n1 = 6
n2 = 9

# resultado con n1 = 6 => n2 + n2 + n2 + n2 + n2 + n2
resultado = 0 # 9, 18, 27, 36, 45, 54,
veces = 1     # 1, 2,  3,  4,  5,  6,   7

while veces <= n1:
    resultado += n2
    veces += 1


print("El resultado de la multiplicación entre ", n1, " y ", n2, " es: ", resultado)
```

## Solución recursiva de la multiplicación a través de sumatorias

```python
def CalcularResultado(veces, n1, n2, resultado):
    print("-----------------------------------------------------------------")
    print("Ambiente #",veces)
    print("Con valores así: n1:", n1, ", n2:", n2, " y resultado temporal: ",resultado)
    print("-----------------------------------------------------------------")
    if veces > n1:
        print(resultado)
    else:
        CalcularResultado(veces + 1, n1, n2, resultado + n2)
```

```python
import sys
sys.setrecursionlimit(10**6) 

n1 = input("Ingrese n1: ")
n2 = input("Ingrese n2: ")
veces = 1
resultado = 0

CalcularResultado(veces, int(n1), int(n2), resultado)
```
